```python
import math
class Triangle:

    def __init__(self, a, b, c):
        self.a = a
        self.b = b
        self.c = c
        if self.a <= 0 or self.b <= 0 or self.c <= 0:
            raise ValueError("Sides of triangle cannot be negative or zero")
        if not self.is_triangle():
            raise ValueError("Sum of any two sides should be greater than third")
            
    def __str__(self):
        return f"Triangle(a = {self.a} , b = {self.b} , c = {self.c})"

    def is_triangle(self):
        return (
            self.a + self.b >= self.c and
            self.b + self.c >= self.a and
            self.c + self.a >= self.b
        )

    def perimeter(self):
        return self.a + self.b + self.c

    def area(self):
        s = self.perimeter() / 2
        return math.sqrt(s*(s-self.a)*(s-self.b)*(s-self.c))
```

# polymorphism in classes

```python
class India:

    def capital(self):
        print("New Delhi is capital of India")

    def language(self):
        print("Hindi is widely spoken in India")
```

```python
class USA:

    def capital(self):
        print("Washington D.C. is capital of USA")

    def language(self):
        print("English is widely spoken in USA")
```

```python
class France:

    def capital(self):
        print("Paris is capital of France")

    def language(self):
        print("French is widely spoken in France")
```

```python
c1 = India()
c1.capital()
c1.language()
```

```python
c2 = USA()
c2.capital()
c2.language()
```

```python
c3 = France()
c3.capital()
c3.language()
```

```python
countries = [India(), USA(), France()]

for i in countries:
    print(type(i))
    i.capital()
    i.language()
    print("===================================")
```

```python
class Rectangle:

    def __init__(self, width: float, height: float):
        self.width = width
        self.height = height

    def perimeter(self):
        return 2 * (self.width + self.height)
    
    def area(self):
        return self.width * self.height
```

```python
import math

print(math.pi)
```

```python
class Circle:

    def __init__(self, radius):
        self.radius = radius 

    def perimeter(self):
        return 2 * math.pi * self.radius
    
    def area(self):
        return math.pi * (self.radius ** 2)
```

```python
t1 = Triangle(a = 3, b = 4, c = 5)
print(t1.perimeter())
print(t1.area())
12
```

```python
c2 = Circle(radius = 20)
print(c2.perimeter())
print(c2.area())
```
