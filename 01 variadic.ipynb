# write a variadis function to add multiple numbers

```python
def add_multiple(*args) -> int | float:
    # Intialize the sum to 0 
    s = 0
    # Add the elements inside args
    for i in args:
        s = s + i
    return s
```

```python
a1 = add_multiple(2, 3)
print(a1)
```

```python
a2 = add_multiple(2, 3, 4, 5, 6)
print(a2)
```

```python
list(range(1, 10))
```

```python
nums = list(range(1, 1001))
print(nums)
a3 = add_multiple(*nums)
print(a3)
```

```python
cart = [
    ("Laptop", 2, 45000),
    ("Mouse", 1, 500),
    ("Keyboard", 2, 1500),
    ("Screen Guard", 3, 800)
]
```

```python
def billing(*args):
    total_bill = 0
    for name, qty, price in args:
        line_total = price * qty
        total_bill = total_bill + line_total
        print(f"Name : {name}, Quantity : {qty}, Price : {price}, Line Total : {line_total}")
        print("="*80)
    return total_bill
```

```python
[('Laptop', 2, 45000),
 ('Mouse', 1, 500),
 ('Keyboard', 2, 1500),
 ('Screen Guard', 3, 800)]
```

```python
b1 = billing(*cart)
print(f"Total Bill : {b1} INR")
```

```python
b2 = billing(
    ("Mobile", 2, 25000),
    ("Cover", 4, 300),
    ("Screen guard", 2, 400)
)
print(f"Total Bill : {b2} INR")
```

# keywords arguments

```python

def function2(**kwargs):
    print(f"Keyword arguments : {kwargs}")
    print(f"Type : {type(kwargs)}")
    for key, value in kwargs.items():
        print(f"Key : {key}, Value : {value}")
```

```python
function2(name="Rahul", age=28)
```

```python
function2(name="Raman", age=32, email="r@test.com", city="Chennai")
```

```python
function2(a=1, b=3, c="Test")
```

#Functions with positional and keyword arguments


```python
def function3(*args, **kwargs):
    print(f"Postional Arguments : {args}")
    print(f"Type Positinal Arguments : {type(args)}")
    for i in args:
        print(i)
    print("="*60)
    print(f"Keyword Arguments : {kwargs}")
    print(f"Type Keyword Arguments : {type(kwargs)}")
    for key, value in kwargs.items():
        print(f"Key : {key}, Value : {value}")
```

```python
function3(1, 2, 3, 4, "Example", "Test", False)
```

```python
function3(a="Rahul", b=23, d=False, id=123)
```

```python
function3(1, 2, 3, "Example", True, a=12, b=13, c="Test", d=False)
```

```python
a = [12, 13, 14, "Example", "Test", False]
b = {"id": 103, "name": "Rahul", "age": 28}
function3(*a, **b)
```
